```python
#training data 

import os
import numpy as np
from PIL import Image
import cv2
import pickle

image_dir ='/home/tilak/Desktop/face/photo'
cas_dir = '/home/tilak/Desktop/face/opencv-master/data/haarcascades/haarcascade_frontalface_alt2.xml'

face_cascade = cv2.CascadeClassifier(cas_dir)
recognizer = cv2.face.LBPHFaceRecognizer_create()


#BASE_DIR = os.path.dirname(os.path.abspath(__file__))
 #os.path.join(BASE_DIR, "images")
current_id = 0
label_ids = {}
y_labels = []
x_train = []

cap = cv2.VideoCapture(0)

for root, dir, files in os.walk(image_dir):
    for file in files:
        if file.endswith("jpg") or file.endswith("png"):
            path = os.path.join(root, file)
            label = os.path.basename(os.path.dirname(path)).replace(" ","_").lower()
            #print(label, path)
            if not label in label_ids:
                label_ids[label] = current_id
                current_id +=1
            id_ = label_ids[label]
            #print(label_ids)
            
            gray =Image.open(path).convert("L") # convert into grayscale
            size =(550, 550)
            gray_resized = gray.resize(size, Image.ANTIALIAS)
            
            image_array = np.array(gray_resized, "uint8")
            #print(image_array)
            faces = face_cascade.detectMultiScale(image_array, scaleFactor=1.5, minNeighbors=5)
            
            for (x, y, w, h) in faces:
                roi = image_array[y:y + h, x:x + h]
                x_train.append(roi)
                y_labels.append(id_)
            

with open("labels.pickle", 'wb') as f:
    pickle.dump(label_ids, f)


recognizer.train(x_train, np.array(y_labels))
recognizer.save("trainner.yml")
            
            
```
